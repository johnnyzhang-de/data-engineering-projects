```python
def normalize_text(text):
    if not isinstance(text,str):
        return None
    text = text.strip()
    return text if text else None
```

```python
def normalize_email(email):
    email = normalize_text(email)
    if not email:
        return None
    email = email.lower()
    if '@' not in email:
        return None
    return email
```

```python
def normalize_phone(phone):
    phone = str(phone)
    if not phone:
        return None
    phone = phone.split('-')
```
